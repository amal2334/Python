# **Data visulization**
- Matplotlib and Seaborn are two of the most popular libraries for data visualization in Python. They allow you to create a wide variety of plots to explore and present your data effectively.
- Let’s go through the most commonly used methods for visualizing data in both libraries.

```python
plt.plot(data['radius_mean'],data['texture_mean'])
plt.title('radius_mean vs texture_mean')
plt.xlabel('radius_mean')
plt.ylabel('texture_mean')
plt.show()
```

```python
plt.plot(data['radius_mean'], data['diagnosis'],color='red')
plt.title('radius_mean vs diagnosis')
plt.xlabel('radius_mean')
plt.ylabel('diagnosis')
plt.show()
```

```python
plt.plot(data['area_mean'], data['diagnosis'],color='red')
plt.title('radius_mean vs diagnosis')
plt.xlabel('radius_mean')
plt.ylabel('diagnosis')
plt.show()
```

```python
plt.scatter(data['radius_mean'], data['diagnosis'])
plt.title('radius_mean vs diagnosis')
plt.xlabel('radius_mean')
plt.ylabel('diagnosis')
plt.show()
```

```python
plt.scatter(data['area_mean'], data['radius_mean'],color='red',label='area_mean')
plt.title('area_mean vs diagnosis')
plt.xlabel('area_mean')
plt.ylabel('diagnosis')
plt.show()
```

```python
data['diagnosis'].value_counts().plot(kind='bar')

```

```python
plt.hist(data['radius_mean'],color='green',bins=10)
plt.xlabel('radius_mean')
plt.ylabel('Frequency')
plt.title('Histogram of radius_mean')
plt.show()
```

```python
plt.boxplot(data['radius_mean'],vert=False)
plt.xlabel('radius_mean')
plt.ylabel('radius_mean')
plt.title('Boxplot of radius_mean')
plt.show()
```

```python
plt.pie(data['diagnosis'].value_counts(),labels=data['diagnosis'].unique(),autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart of diagnosis')
```

```python
# now let's test the column radius means even thoh it is numerical
plt.pie(data['radius_mean'].value_counts(),labels=data['radius_mean'].unique(),autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie chart of radius_mean')
```

```python
#seaborn styling
import seaborn as sns
sns.set_style('darkgrid')


sns.lineplot(x=data['radius_mean'],y=data['area_mean'],color='red')
plt.title('radius_mean vs area_mean')
plt.xlabel('radius_mean')
plt.ylabel('area_mean')
plt.show()
```

```python
sns.lineplot(x=data['radius_mean'],y=data['texture_mean'],color='red',label='texture_mean')
plt.title('radius_mean vs area_mean')
plt.legend()
plt.xlabel('radius_mean')
plt.ylabel('area_mean')
plt.show()
```

```python
sns.scatterplot(x=data['radius_mean'],y=data['area_mean'],color='green',data=data)
plt.title('radius_mean vs area_mean')
plt.xlabel('radius_mean')
plt.ylabel('area_mean')
plt.legend()
plt.show()
```
